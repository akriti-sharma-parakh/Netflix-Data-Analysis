PRE-PROCESSING of DATA


Unnesting 'cast', 'director' and 'country'

```python
def unnest(df,column):
  return df.explode(column).reset_index(drop = True)

```

```python
df_original = df.copy(deep = True)
```

```python
df_original
```

```python
df = df_original.copy(deep=True)
for field in ['cast', 'director','country']:
  df[field] = df[field].str.split(', ')
  df = unnest(df,field)
df
```
